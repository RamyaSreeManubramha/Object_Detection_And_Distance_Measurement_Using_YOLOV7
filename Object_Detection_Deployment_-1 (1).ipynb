```python
import logging

# Suppress Ultralytics logging
logging.getLogger('ultralytics').setLevel(logging.WARNING)
```

```python
from object_detection import detect_objects
import cv2
import gradio as gr
import numpy as np

# Define the Gradio interface
iface = gr.Interface(
    fn=detect_objects,  # The function to process the input
    inputs=gr.Image(type="numpy", label="Upload Image"),  # Input component
    outputs=gr.Image(type="numpy", label="Detected Objects"),  # Output component
    title="Object Detector: Chairs, Tables, and Trashcans",
    description="Upload an image, and the model will detect Chairs, Tables, and Trashcans, drawing bounding boxes around them with distinct colors.",
    allow_flagging="never"
)

# Launch the Gradio app
iface.launch(share=True, debug=True)  # Use share=True for Colab to generate a public link
```
